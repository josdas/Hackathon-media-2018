```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import pandas_profiling
import numpy as np
import imageio
```

```python
TEST_VIDEO = '../../data/train/633012_5.mp4'
```

```python
def get_reader(path, frame_rate=1):
    for i, frame in enumerate(imageio.get_reader(path).iter_data()):
        if i % frame_rate == 0:
            yield frame
```

```python
tt = get_reader(TEST_VIDEO, frame_rate=10)
```
